```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

#imarray = numpy.random.rand(100,100,3) * 255
#im = Image.fromarray(imarray.astype('uint8')).convert('RGBA')
#im.save("result_image.png")

# user input parameters
width, height, a, b, delta, user_color, image_path = input("Enter: width, height, a, b, delta, color, image path").split()

def lissajous_curve(width, height, a, b, delta, user_color, image_path):
    # time (start, stop, number of steps)
    t = np.linspace(0, np.pi * 2, 10000)
    # x, y variables
    x = np.sin(a * t + np.pi / delta)
    y = np.sin(b * t)
    # render graph
    fig, ax = plt.subplots()
    ax.plot(x, y, color = user_color)
    # set transparent background and remove plot axes
    fig.patch.set_visible(False)
    ax.axis('off')
    # save plot
    fig.savefig("plot.png")
    # Pillow: open, resize and save
    image = Image.open("plot.png")
    resized_image = image.resize((width, height))
    resized_image.save(f"{image_path}/result_image.png")
    
lissajous_curve(int(width), int(height), int(a), int(b), int(delta), str(user_color), image_path)
```
